```python
plt.figure(figsize=(10,5))
sns.histplot(data['Age'], bins= 10)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.histplot(data['EstimatedSalary'], bins = 10)
plt.title('EstimatedSalary Distribution')
plt.xlabel('EstimatedSalary')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.histplot(data['Purchased'], bins = 10)
plt.title('Purchased Distribution')
plt.xlabel('Purchased')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize = (12,16))
plt.subplot(321)
plt.hist(data['Age'], bins = 20, color = 'orange', alpha = 1)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.subplot(322)
plt.hist(data['EstimatedSalary'], bins = 20, color = 'blue', alpha = 1)
plt.title('EstimatedSalary Distribution')
plt.xlabel('EstimatedSalary')
plt.subplot(323)
plt.hist(data['Purchased'], bins = 20, color = 'black', alpha = 1)
plt.title('Purchased Distribution')
plt.xlabel('Purchaed')
plt.show()
```

```python
sns.boxplot(x = 'Purchased', y = 'Age', data = data)
plt.title('Age vs Purchased')
plt.show()
```

```python
sns.boxplot(x = 'Purchased', y = 'EstimatedSalary', data = data)
plt.title('EstimatedSalary vs Purchased')
plt.show()
```

```python
sns.boxplot(data = data[['Age', 'EstimatedSalary']])
plt.title('Boxplot of Age abd EstimatedSalary')
plt.show()
```

```python
sns.scatterplot(x = 'Age', y = 'Purchased', data=data)
plt.title('Age Distribution vs. Purchased')
```

```python
sns.scatterplot(x = 'EstimatedSalary', y = 'Purchased', data= data)
plt.title('EstimatedSalary Distribution vs. Purchased')
```

```python
plt.figure(figsize = (12,4))
plt.subplot(121)
plt.scatter(data['Age'], data['Purchased'],color = 'orange', alpha = 1)
plt.title('Age vs. Purchased')
plt.xlabel('Age Distribution')
plt.ylabel('Purchased')
plt.subplot(122)
plt.scatter(data['EstimatedSalary'], data['Purchased'], color = 'blue', alpha =1)
plt.title('EstimatedSalary vs. Purchased')
plt.xlabel('EstimatedSalary Distribution')
plt.ylabel('Purchased')
```

```python
plt.figure(figsize = (8,6))
sns.heatmap(data.corr(), annot = True, cmap = 'viridis')
plt.title('Correlation Heatmap')
plt.show()
```
